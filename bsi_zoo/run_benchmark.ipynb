```python
import os
import numpy as np
import pandas as pd

path = "/home/anuja/tj/bdsg/BSI-Zoo/bsi_zoo/data/free4"
files = os.listdir(path)
benchmark_files_free =  [i for i in files if 'spatialCV' not in i]
benchmark_files_free_with_spatial_cv = [i for i in files if 'spatialCV' in i]


eloreta_path = "/home/anuja/tj/bdsg/BSI-Zoo/bsi_zoo/data/eloreta"
files = os.listdir(eloreta_path)
benchmark_files_eloreta =  [i for i in files if 'benchmark' not in i]

dfs_free = [pd.read_pickle(f'{path}/{file}') for file in benchmark_files_free]
df_results_free = pd.concat(dfs_free)
df_results_free['alpha'] = 1-df_results_free['alpha'].astype(float)



dfs_free_with_spatial_cv = [pd.read_pickle(f'{path}/{file}') for file in benchmark_files_free_with_spatial_cv]
eloreta_dfs = [pd.read_pickle(f'{eloreta_path}/{file}') for file in benchmark_files_eloreta]
df_results_free_with_spatial_cv = pd.concat(dfs_free_with_spatial_cv)
df_results_eloreta = pd.concat(eloreta_dfs)

# join with eloreta
df_results_free_with_spatial_cv = pd.concat([df_results_eloreta, df_results_free_with_spatial_cv])
df_results_free_with_spatial_cv['alpha'] = 1-df_results_free_with_spatial_cv['alpha'].astype(float)
```

```python
# save df_results_free as .mat file
import scipy.io
Output = {}

scipy.io.savemat('free_spatialCV.mat', df_results_free_with_spatial_cv.to_dict('list'))
```
